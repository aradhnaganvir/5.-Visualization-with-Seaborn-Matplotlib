```python
data = pd.read_excel('olympics_data.xlsx')
data.columns = ['country','games_played','gold','silver','bronze','total']

def rating(medals):
    if medals >= 500:
        return 'A'
    elif medals < 500 and medals >= 300:
        return 'B'
    elif medals < 300 and medals >= 100:
        return 'C'
    elif medals < 50 and medals >= 100:
        return 'D'
    elif medals < 10 and medals >= 50:
        return 'E'
    elif medals < 10 and medals >= 1:
        return 'F'
    else:
        return 'G'
    
data['rating'] = data['total'].apply(rating)
data.head(3)
```

```python
mean_medal = data[['gold','silver','bronze','rating']].groupby(by='rating').mean()
```

```python
medal_total = data[['total','rating']].groupby(by='rating').mean()
```

```python
medal_total.plot(kind='bar')
```
